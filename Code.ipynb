```python
import os
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch import einsum
from einops import rearrange
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader, random_split
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score
from PIL import Image
import matplotlib.pyplot as plt
from tqdm import tqdm  # progress bar
import pydensecrf.densecrf as dcrf
# ----------------------------
# 1. Dataset Loading with Caching and Progress Indicator
# ----------------------------
class SegmentationDataset(Dataset):
    def __init__(self, image_paths, mask_paths, transform_img=None, transform_mask=None, cache=False):
        assert len(image_paths) == len(mask_paths), "Mismatch between number of images and masks!"
        self.image_paths = image_paths
        self.mask_paths = mask_paths
        self.transform_img = transform_img
        self.transform_mask = transform_mask
        self.cache = cache
        self.cached_data = {} if cache else None

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        if self.cache and idx in self.cached_data:
            return self.cached_data[idx]
        image = Image.open(self.image_paths[idx]).convert("RGB")
        mask = Image.open(self.mask_paths[idx]).convert("L")
        if self.transform_img:
            image = self.transform_img(image)
        if self.transform_mask:
            mask = self.transform_mask(mask)
        if self.cache:
            self.cached_data[idx] = (image, mask)
        return image, mask

def get_image_and_mask_paths(dataset_root, dataset_names):
    image_paths = []
    mask_paths = []
    for ds in dataset_names:
        img_dir = os.path.join(dataset_root, ds, "images")
        mask_dir = os.path.join(dataset_root, ds, "masks")
        if os.path.exists(img_dir) and os.path.exists(mask_dir):
            files = sorted(os.listdir(img_dir))
            for file in tqdm(files, desc=f"Loading images for {ds}"):
                if file.endswith(('.jpg', '.png')):
                    image_paths.append(os.path.join(img_dir, file))
                    mask_file = os.path.splitext(file)[0] + ".png"
                    mask_paths.append(os.path.join(mask_dir, mask_file))
        else:
            print(f"Skipping dataset {ds}: directory not found.")
    return image_paths, mask_paths

# ----------------------------
# 2. Model Components
# ----------------------------
# Separable Convolution for efficient feature extraction
class SeparableConv2d(nn.Module):
    def __init__(self, in_channels, out_channels, kernel_size=3, stride=1, padding=1):
        super().__init__()
        self.depthwise = nn.Conv2d(in_channels, in_channels, kernel_size,
                                   stride=stride, padding=padding, groups=in_channels)
        self.pointwise = nn.Conv2d(in_channels, out_channels, 1)

    def forward(self, x):
        return self.pointwise(self.depthwise(x))

# Basic MultiResBlock using separable convolutions
class MultiResBlock(nn.Module):
    def __init__(self, in_channels, out_channels):
        super().__init__()
        self.conv1 = SeparableConv2d(in_channels, out_channels)
        self.conv2 = SeparableConv2d(out_channels, out_channels)
        self.conv3 = SeparableConv2d(out_channels, out_channels)
        self.res_conv = nn.Conv2d(in_channels, out_channels, 1) if in_channels != out_channels else nn.Identity()
        self.bn = nn.BatchNorm2d(out_channels)

    def forward(self, x):
        residual = self.res_conv(x)
        x = F.relu(self.conv1(x))
        x = F.relu(self.conv2(x))
        x = F.relu(self.conv3(x))
        return self.bn(x + residual)

# ----------------------------
# CBAM: Convolutional Block Attention Module
# ----------------------------
class ChannelAttention(nn.Module):
    def __init__(self, in_planes, reduction_ratio=16):
        super(ChannelAttention, self).__init__()
        self.avg_pool = nn.AdaptiveAvgPool2d(1)
        self.max_pool = nn.AdaptiveMaxPool2d(1)
        self.fc = nn.Sequential(
            nn.Conv2d(in_planes, in_planes // reduction_ratio, 1, bias=False),
            nn.ReLU(),
            nn.Conv2d(in_planes // reduction_ratio, in_planes, 1, bias=False)
        )
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        avg_out = self.fc(self.avg_pool(x))
        max_out = self.fc(self.max_pool(x))
        out = avg_out + max_out
        return self.sigmoid(out)

class SpatialAttention(nn.Module):
    def __init__(self, kernel_size=7):
        super(SpatialAttention, self).__init__()
        self.conv = nn.Conv2d(2, 1, kernel_size, padding=(kernel_size - 1) // 2, bias=False)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        avg_out = torch.mean(x, dim=1, keepdim=True)
        max_out, _ = torch.max(x, dim=1, keepdim=True)
        x_cat = torch.cat([avg_out, max_out], dim=1)
        x_out = self.conv(x_cat)
        return self.sigmoid(x_out)

class CBAM(nn.Module):
    def __init__(self, in_planes, reduction_ratio=16, kernel_size=7):
        super(CBAM, self).__init__()
        self.channel_attention = ChannelAttention(in_planes, reduction_ratio)
        self.spatial_attention = SpatialAttention(kernel_size)

    def forward(self, x):
        out = x * self.channel_attention(x)
        out = out * self.spatial_attention(out)
        return out

# Enhanced MultiResBlock with integrated CBAM module.
class MultiResBlock_CBAM(nn.Module):
    def __init__(self, in_channels, out_channels, use_cbam=True):
        super().__init__()
        self.multi_res = MultiResBlock(in_channels, out_channels)
        self.use_cbam = use_cbam
        if use_cbam:
            self.cbam = CBAM(out_channels)

    def forward(self, x):
        x = self.multi_res(x)
        if self.use_cbam:
            x = self.cbam(x)
        return x

# PAM module for self-attention
class PAM_Module(nn.Module):
    def __init__(self, in_dim):
        super().__init__()
        self.query_conv = nn.Conv2d(in_dim, in_dim // 8, 1)
        self.key_conv = nn.Conv2d(in_dim, in_dim // 8, 1)
        self.value_conv = nn.Conv2d(in_dim, in_dim, 1)
        self.gamma = nn.Parameter(torch.zeros(1))
        self.softmax = nn.Softmax(dim=-1)

    def forward(self, x):
        batch, C, H, W = x.size()
        proj_query = self.query_conv(x).view(batch, -1, H * W).permute(0, 2, 1)
        proj_key = self.key_conv(x).view(batch, -1, H * W)
        energy = torch.bmm(proj_query, proj_key)
        attention = self.softmax(energy)
        proj_value = self.value_conv(x).view(batch, -1, H * W)
        out = torch.bmm(proj_value, attention.permute(0, 2, 1))
        out = out.view(batch, C, H, W)
        return self.gamma * out + x

# SPAM module with adaptive patch extraction
class SPAM(nn.Module):
    def __init__(self, in_dim, patch_size=4, adaptive=False, patch_divisor=32):
        """
        Args:
            in_dim (int): Number of input channels.
            patch_size (int): Fixed patch size when adaptive is False.
            adaptive (bool): Whether to compute patch size adaptively.
            patch_divisor (int): Divisor for adaptive patch size computation.
        """
        super().__init__()
        self.patch_size = patch_size
        self.adaptive = adaptive
        self.patch_divisor = patch_divisor
        self.pam = PAM_Module(in_dim)
        self.conv = nn.Conv2d(in_dim, in_dim, 1)

    def forward(self, B):
        B1 = B
        B2 = B
        batch, C, H, W = B1.shape
        # Compute patch size adaptively if enabled
        if self.adaptive:
            N = max(2, H // self.patch_divisor)
        else:
            N = self.patch_size
        patches1 = rearrange(B1, 'b c (h p1) (w p2) -> b (h w) c p1 p2', p1=N, p2=N)
        patches2 = rearrange(B2, 'b c (h p1) (w p2) -> b (h w) c p1 p2', p1=N, p2=N)
        patches1_norm = F.normalize(patches1.flatten(3), dim=-1)
        patches2_norm = F.normalize(patches2.flatten(3), dim=-1)
        sim_matrix = einsum('b n p, b m p -> b n m', patches1_norm.mean(2), patches2_norm.mean(2))
        mask = torch.eye(sim_matrix.size(1), device=sim_matrix.device).unsqueeze(0)
        sim_matrix = sim_matrix * (1 - mask)
        sim_scores = sim_matrix.mean(dim=-1)
        sim_map = rearrange(sim_scores, 'b (h w) -> b 1 h w', h=H//N, w=W//N)
        sim_map = F.interpolate(sim_map, size=(H, W), mode='nearest')
        Bout = self.pam(B)
        return Bout * sim_map

# ----------------------------
# Enhanced MultiResUNet with SPAM, Adaptive Patch Extraction, and CBAM
# ----------------------------
class MultiResUNet_SPAM_CBAM(nn.Module):
    def __init__(self, in_channels=3, out_channels=1):
        super().__init__()
        # Encoder with CBAM-enhanced blocks
        self.encoder1 = MultiResBlock_CBAM(in_channels, 64)
        self.down1 = nn.Conv2d(64, 64, kernel_size=2, stride=2)

        self.encoder2 = MultiResBlock_CBAM(64, 128)
        self.down2 = nn.Conv2d(128, 128, kernel_size=2, stride=2)

        self.encoder3 = MultiResBlock_CBAM(128, 256)
        self.down3 = nn.Conv2d(256, 256, kernel_size=2, stride=2)

        # Bottleneck with SPAM (adaptive patch extraction)
        self.bottleneck = MultiResBlock_CBAM(256, 512)
        self.spam = SPAM(512, patch_size=4, adaptive=True, patch_divisor=32)

        # Decoder (optionally CBAM can also be added here)
        self.up3 = nn.ConvTranspose2d(512, 256, kernel_size=2, stride=2)
        self.decoder3 = MultiResBlock_CBAM(512, 256)

        self.up2 = nn.ConvTranspose2d(256, 128, kernel_size=2, stride=2)
        self.decoder2 = MultiResBlock_CBAM(256, 128)

        self.up1 = nn.ConvTranspose2d(128, 64, kernel_size=2, stride=2)
        self.decoder1 = MultiResBlock_CBAM(128, 64)

        # Final layer
        self.final = nn.Conv2d(64, out_channels, kernel_size=1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        e1 = self.encoder1(x)
        d1 = self.down1(e1)

        e2 = self.encoder2(d1)
        d2 = self.down2(e2)

        e3 = self.encoder3(d2)
        d3 = self.down3(e3)

        b = self.bottleneck(d3)
        b = self.spam(b)

        up3 = self.up3(b)
        up3 = torch.cat([up3, e3], dim=1)
        d3 = self.decoder3(up3)

        up2 = self.up2(d3)
        up2 = torch.cat([up2, e2], dim=1)
        d2 = self.decoder2(up2)

        up1 = self.up1(d2)
        up1 = torch.cat([up1, e1], dim=1)
        d1 = self.decoder1(up1)

        out = self.sigmoid(self.final(d1))
        return out

# ----------------------------
# 3. Loss Function: Composite Loss (BCE + IoU)
# ----------------------------
class CompositeLoss(nn.Module):
    def __init__(self):
        super().__init__()
        self.bce = nn.BCELoss()

    def iou_loss(self, pred, target):
        intersection = (pred * target).sum()
        union = pred.sum() + target.sum() - intersection
        return 1 - (intersection + 1e-6) / (union + 1e-6)

    def forward(self, pred, target):
        bce_loss = self.bce(pred, target)
        iou = self.iou_loss(pred, target)
        return bce_loss + iou

# ----------------------------
# 4. Training, Evaluation, and Metrics
# ----------------------------
def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=30, device="cuda"):
    model.to(device)
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for imgs, masks in train_loader:
            imgs = imgs.to(device)
            masks = masks.to(device)
            optimizer.zero_grad()
            outputs = model(imgs)
            loss = criterion(outputs, masks)
            loss.backward()
            optimizer.step()
            running_loss += loss.item() * imgs.size(0)
        epoch_loss = running_loss / len(train_loader.dataset)

        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for imgs, masks in val_loader:
                imgs = imgs.to(device)
                masks = masks.to(device)
                outputs = model(imgs)
                loss = criterion(outputs, masks)
                val_loss += loss.item() * imgs.size(0)
        val_loss /= len(val_loader.dataset)
        print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {epoch_loss:.4f}, Val Loss: {val_loss:.4f}")
    return model

def evaluate_model(model, test_loader, device="cuda"):
    model.eval()
    preds = []
    trues = []
    raw_probs = []
    with torch.no_grad():
        for imgs, masks in test_loader:
            imgs = imgs.to(device)
            outputs = model(imgs)
            raw_probs.append(outputs.cpu())
            preds.append((outputs > 0.5).float().cpu())
            trues.append(masks)
    preds = torch.cat(preds, dim=0).numpy()
    trues = torch.cat(trues, dim=0).numpy()
    raw_probs = torch.cat(raw_probs, dim=0).numpy()
    preds_binary = (raw_probs > 0.5).astype(np.uint8).flatten()
    trues_binary = (trues > 0.5).astype(np.uint8).flatten()
    return trues_binary, preds_binary, raw_probs

def compute_metrics(y_true, y_pred):
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    accuracy = accuracy_score(y_true, y_pred)
    return accuracy, precision, recall, f1

# ----------------------------
# Helper: Compute IoU for single image (for comparison)
# ----------------------------
def compute_iou_single(gt, pred):
    gt = gt.astype(np.uint8)
    pred = pred.astype(np.uint8)
    intersection = np.logical_and(gt, pred).sum()
    union = np.logical_or(gt, pred).sum()
    if union == 0:
        return 1.0
    return intersection / union

# ----------------------------
# Visualization: Compare prediction before and after CRF refinement
# ----------------------------
def visualize_prediction(image, mask_true, mask_pred, refined_mask=None, iou_pred=None, iou_refined=None):
    image = image.permute(1, 2, 0).numpy()
    mask_true = mask_true.squeeze().numpy()
    mask_pred = mask_pred.squeeze().numpy()
    plt.figure(figsize=(18, 4))
    plt.subplot(1, 5, 1)
    plt.imshow(image)
    plt.title("Input Image")
    plt.subplot(1, 5, 2)
    plt.imshow(mask_true, cmap='gray')
    plt.title("Ground Truth")
    plt.subplot(1, 5, 3)
    plt.imshow(mask_pred, cmap='gray')
    title_pred = "Prediction"
    if iou_pred is not None:
        title_pred += f"\nIoU: {iou_pred:.4f}"
    plt.title(title_pred)
    if refined_mask is not None:
        plt.subplot(1, 5, 4)
        plt.imshow(refined_mask, cmap='gray')
        title_ref = "CRF Refined"
        if iou_refined is not None:
            title_ref += f"\nIoU: {iou_refined:.4f}"
        plt.title(title_ref)
    plt.subplot(1, 5, 5)
    # Show difference between CRF refined and prediction
    diff = np.abs((mask_pred > 0.5).astype(np.uint8) - (refined_mask > 0.5).astype(np.uint8))
    plt.imshow(diff, cmap='hot')
    plt.title("Difference Map")
    plt.show()

# ----------------------------
# 5. Conditional Random Field (CRF) Refinement
# ----------------------------
def apply_crf(image, output_prob):
    """
    Apply CRF refinement to the segmentation mask.
    Args:
        image (numpy.ndarray): Input RGB image (H, W, 3) with values [0,255].
        output_prob (numpy.ndarray): Predicted probability mask (H, W) in [0,1].
    Returns:
        numpy.ndarray: Refined segmentation mask.
    """
    try:
        import pydensecrf.densecrf as dcrf
        from pydensecrf.utils import unary_from_softmax
    except ImportError:
        raise ImportError("pydensecrf is required. Install via 'pip install pydensecrf'.")

    # Ensure the input image is C-contiguous
    image = np.ascontiguousarray(image)

    H, W = output_prob.shape
    d = dcrf.DenseCRF2D(W, H, 2)
    unary = unary_from_softmax(np.stack([1 - output_prob, output_prob], axis=0))
    d.setUnaryEnergy(unary)
    d.addPairwiseGaussian(sxy=3, compat=3)
    d.addPairwiseBilateral(sxy=80, srgb=13, rgbim=image, compat=10)
    Q = np.array(d.inference(5))  # Convert Q to a NumPy array
    refined_mask = Q[1].reshape((H, W))  # Access the second class probability and reshape

    return refined_mask


# ----------------------------
# 6. Main: Prepare Data, Train, Evaluate, Visualize, and Compare Outputs
# ----------------------------
if __name__ == "__main__":
    # Define dataset root and names (update path as needed)
    dataset_root = "/content/drive/MyDrive/dataset pro"
    dataset_names = ["CASIA2", "CoMoFoD_small_v2", "MICC_F600", "Coverage"]

    # Define transforms for images and masks
    transform_img = transforms.Compose([
        transforms.Resize((256, 256)),
        transforms.ToTensor(),
    ])
    transform_mask = transforms.Compose([
        transforms.Resize((256, 256)),
        transforms.ToTensor(),
    ])

    # Gather file paths with progress indicator
    img_paths, mask_paths = get_image_and_mask_paths(dataset_root, dataset_names)
    if not img_paths:
        raise ValueError("No valid data found. Check dataset directories.")

    # Create dataset with caching enabled (if memory permits)
    dataset = SegmentationDataset(img_paths, mask_paths, transform_img, transform_mask, cache=True)

    # Split dataset into train (70%), val (20%), and test (10%)
    total_size = len(dataset)
    train_size = int(0.7 * total_size)
    val_size = int(0.2 * total_size)
    test_size = total_size - train_size - val_size
    train_dataset, val_dataset, test_dataset = random_split(dataset, [train_size, val_size, test_size])

    # DataLoader parameters
    device = "cuda" if torch.cuda.is_available() else "cpu"
    train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, num_workers=4, pin_memory=True)
    val_loader = DataLoader(val_dataset, batch_size=8, shuffle=False, num_workers=4, pin_memory=True)
    test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False, num_workers=4, pin_memory=True)

    print(f"Total samples: {total_size} | Train: {train_size} | Val: {val_size} | Test: {test_size}")

    # Initialize model, loss, and optimizer
    model = MultiResUNet_SPAM_CBAM(in_channels=3, out_channels=1).to(device)
    criterion = CompositeLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

    # Train the model
    model = train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs=30, device=device)

    # Evaluate the model (overall metrics)
    y_true, y_pred, raw_probs = evaluate_model(model, test_loader, device=device)
    accuracy, precision, recall, f1 = compute_metrics(y_true, y_pred)
    print(f"Overall Metrics:\nAccuracy: {accuracy:.4f}\nPrecision: {precision:.4f}\nRecall: {recall:.4f}\nF1 Score: {f1:.4f}")

    # Visualize and compare predictions for first 5 test samples:
    model.eval()
    with torch.no_grad():
        for i in range(min(5, len(test_dataset))):
            img, m = test_dataset[i]
            img_input = img.unsqueeze(0).to(device)
            pred_prob = model(img_input)  # probability output
            pred = (pred_prob > 0.5).float().cpu()

            # Prepare image for CRF: convert tensor image to numpy uint8 in [0,255]
            img_np = (img.permute(1, 2, 0).numpy() * 255).astype(np.uint8)
            pred_prob_np = pred_prob.squeeze().cpu().numpy()

            # Apply CRF refinementY
            refined_mask = apply_crf(img_np, pred_prob_np)

            # Compute per-sample IoU for predicted and CRF refined outputs:
            gt_np = (m.squeeze().cpu().numpy() > 0.5).astype(np.uint8)
            pred_np = (pred.squeeze().numpy() > 0.5).astype(np.uint8)
            refined_np = (refined_mask > 0.5).astype(np.uint8)
            iou_pred = compute_iou_single(gt_np, pred_np)
            iou_refined = compute_iou_single(gt_np, refined_np)
            print(f"Sample {i}: IoU before CRF: {iou_pred:.4f}, after CRF: {iou_refined:.4f}")

            visualize_prediction(img.cpu(), m, pred.squeeze(), refined_mask=refined_mask,
                                   iou_pred=iou_pred, iou_refined=iou_refined)
```
